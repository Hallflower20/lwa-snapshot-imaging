```python
import numpy as np
import glob
import astropy
from astropy.io import fits
import os
from matplotlib import colors
import matplotlib.pyplot as plt
```

```python
def plot_data(obj,percentile = [0, 99.9]):
    plt.imshow(obj, origin='lower', cmap='gray', norm=colors.LogNorm(), vmin = 0, vmax = 70)
```

```python
time_stamp_n = "night"
data_location = "/lustre/xhall/test_file/"
file_n = time_stamp_n + "-image.fits"

time_stamp_d = "day"
data_location = "/lustre/xhall/test_file/"
file_d = time_stamp_d + "-image.fits"
```

```python
image_n = fits.open(os.path.join(data_location, file_n))[0].data[0][0]
image_d = fits.open(os.path.join(data_location, file_d))[0].data[0][0]
#image_flag = fits.open(os.path.join(data_location, file_flag))[0].data[0][0]
```

```python
plt.figure(figsize = (12,10))
plt.imshow(image_n, origin='lower', cmap='gray', vmin = 0, vmax = 70)
plt.title(time_stamp_n)
plt.colorbar()
```

```python
plt.figure(figsize = (12,10))
plt.imshow(image_d, origin='lower', cmap='gray', vmin = 0, vmax = 20)
plt.title(time_stamp_d)
plt.colorbar()
```
